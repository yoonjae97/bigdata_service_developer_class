```python
from selenium import webdriver as wb
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from tqdm import tqdm

import time
import pandas as pd
```

```python
driver = wb.Chrome()
driver.get('https://www.hsd.co.kr/')

driver.maximize_window()

pop_up = driver.find_element('xpath', '//*[@id="pop"]/div[1]/div/div[2]/span[1]')
if pop_up:
    pop_up.click()
    
menu = driver.find_element('xpath', '//*[@id="gnb"]/div[2]/ul/li[3]/p/a')
menu_all = driver.find_element(By.XPATH, '//*[@id="gnb"]/div[2]/ul/li[3]/div/ul/li[1]/a')
```

```python
action = ActionChains(driver)

# 첫번째 관문, 메뉴로 이동
# menu 버튼 위로 커서 이동
action.move_to_element(menu)
time.sleep(1)

# 두번째 관문, 전체 메뉴 클릭
action.click(menu_all)

# 해당 작업 수행
action.perform()
```

```python
title_lst = []
price_lst = []

more = driver.find_element('xpath', '//*[@id="btn_more"]/span/a')

try:
    while  True:
        time.sleep(1)
        more.click()
except:
    pass

titles = driver.find_elements(By.CSS_SELECTOR, 'h4.h.fz_03')
# class="blind"로 인하여 '가격:' 추출이 되지않음
prices = driver.find_elements(By.CSS_SELECTOR, 'div.item-price')

for i in tqdm(range(len(titles))):
    title_lst.append(titles[i].text)
    price_lst.append(prices[i].text)

driver.quit()    
    
hansot_df = pd.DataFrame({'menu':title_lst, 'price':price_lst})
hansot_df.drop_duplicates(inplace=True)
# hansot_df.to_csv("hansot_menu.csv", encoding='euc-kr', index=False)
hansot_df.to_excel("hansot_menu.xlsx", encoding='euc-kr', index=False)
```

### 10. Selenium_practice5_extra_kakaomap

#### kakao map을 이용한 이수 맛집 검색

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

#### beautifulsoup과 selenium 같이 활용
##### 이유
- selenium은 브라우저 제어 등 자동화 특화
- selenium 자체로 html 파싱, 데이터 추출은 한계가 존재
- beautifulsoup을 사용해서 상호보완

```python
driver = wb.Chrome()
driver.get('https://map.kakao.com/')
driver.maximize_window()
```

```python
click_blue_popup = driver.find_element(By.CSS_SELECTOR, 'body > div.coach_layer.coach_layer_type1 > div > div > div > span')
click_blue_popup.click()
```

```python
search = driver.find_element(By.XPATH, '//*[@id="search.keyword.query"]')
search.send_keys('이수역 맛집')
search.send_keys(Keys.ENTER)
time.sleep(2)
```

```python
soup = bs(driver.page_source, 'lxml')
titles = soup.select('a.link_name')
addresses = soup.select('div.addr > p:nth-child(1)')
```

```python
loc_more = driver.find_element('xpath', '//*[@id="info.search.place.more"]')
loc_more.click()
time.sleep(2)
```

```python
page_buttons = driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a')
```

```python
titles_list = []
addresses_list = []

try:
    while True:
        for i in tqdm(range(1, 6)):
            
            if (i < 5):
                
                page_buttons[i].click()
                time.sleep(2)
                
                soup = bs(driver.page_source, 'lxml')
                titles = soup.select('a.link_name')
                addresses = soup.select('div.addr > p:nth-child(1)')
                
                for i in range(len(titles)):            
                    titles_list.append(titles[i].text)
                    addresses_list.append(addresses[i].text)
            
            else:
                click_next = driver.find_element(By.XPATH, '//*[@id="info.search.page.next"]')
                click_next.click()
                time.sleep(3)
                
except:
    print('수집 완료')
```

```python
titles_list = []
addresses_list = []

try:
    while True:
        
        for i in page_buttons:
            i.click()
            time.sleep(2)
                
            soup = bs(driver.page_source, 'lxml')
            titles = soup.select('a.link_name')
            addresses = soup.select('div.addr > p:nth-child(1)')

            for i in range(len(titles)):            
                titles_list.append(titles[i].text)
                addresses_list.append(addresses[i].text)
            
        click_next = driver.find_element(By.XPATH, '//*[@id="info.search.page.next"]')
        click_next.click()
        time.sleep(3)
        page_buttons = driver.find_elements(By.CSS_SELECTOR, '#info\.search\.page > div > a')
                
except:
    print('수집 완료')
```

```python
kakaodata = pd.DataFrame({'식당 이름':titles_list, '식당 주소':addresses_list})
kakaodata.to_csv('kakao_isu_data.csv', encoding='utf-8', index=False)
```
