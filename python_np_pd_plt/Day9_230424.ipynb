## matplotlib (mathdatas plot library)
- 데이터 분석 결과를 그래프나 차트로 시각화 해주기 위한 라이브러리
- 분석 결과를 수치만으로 설명하는 것보다 시각적인 도구들을 활용하면 훨씬 더 직관적으로 이해가 가능하고 타인에게 전달하기도 용이함

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 플롯 차트(plot chart) : 꺾은선 그래프

```python
# x값 없이 선 그래프 그려보기
y = [2, 4, 6]

# plot : 꺾은선 그래프 출력 함수
plt.plot(y)
# show : 차트를 화면에 표시해주는 함수(쥬피터 노트북에서는 안써도 인식됨)
plt.show()
```

```python
# x, y값 지정하여 선 그래프 그려보기
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

plt.plot(x, y)
plt.show()
```

#### 차트 style 설정

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

plt.plot(x, y, marker='o', c='red', lw=0.5, ls='-.', mec='yellow', ms=5, mfc='blue')
plt.show()
```

#### Linestyle

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

plt.plot(x, y, ls='-.')
plt.show()
```

#### Line width

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

# lw는 숫자로 설정 가능
plt.plot(x, y, ls='-.', lw=3)
plt.show()
```

#### Marker

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

plt.plot(x, y, ls='-.', lw=0.5, marker='D')
plt.show()
```

#### Line color

```python
x = [1, 2, 3, 4, 5]
y = [1, 5, 7, 8, 6]

plt.plot(x, y, ls='-.', lw=0.5, marker='D', c='c')
plt.show()
```

#### 차트 옵션

#### x축, y축 범위 설정

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw=4, marker='o', mec='g', mfc='r', mew=5, ms=15)
plt.xlim(-2, 10)
plt.ylim(-5, 12)
plt.show()
```

#### tick 설정
- 플롯이나 차트에서 축상의 위치 표시 지점을 tick이라고 함
- tick 위에 써진 숫자 혹은 문자는  tick label이라고 함
- tick 위치나 label은 matplotlib에서 자동으로 정해주지만 수동 설정을 하고 싶다면 xticks, yticks 명령 사용

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw=4, marker='o', mec='g', mfc='r', mew=5, ms=15)
# ctrl + d : 한줄씩 삭제 단축키
plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])

plt.show()
```

#### 그리드(grid)
- 그림의 틱 위치를 잘 보여주기 위해 중간에 그리드 선(grid line)을 그어주는 것

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw=4, marker='o', mec='g', mfc='r', mew=5, ms=15)

plt.xticks([0, 3, 6])
plt.yticks([1, 5, 9])
plt.grid()
plt.show()
```

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw=4, marker='o', mec='g', mfc='r', mew=5, ms=15)

plt.grid()
plt.show()

# tick이 있는 지점에만 grid가 생성됨
```

#### x축, y축 제목 및 그래프 제목

```python
x = np.arange(7)
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, ls='--', lw=4, marker='o', mec='g', mfc='r', mew=5, ms=15)

plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0, labelpad=20)
plt.title('title')

plt.grid()
plt.show()
```

#### 범례 설정 + 여러개 플롯 한꺼번에 그리기

```python
x = np.arange(4)

y1 = [1, 2, 3, 4]
y2 = [2, 4, 6, 8]
y3 = [3, 6, 9, 12]

plt.plot(x, y1, label='y1')
plt.plot(x, y2, label='y2')
plt.plot(x, y3, label='y3')

plt.xlabel('X axis')
plt.ylabel('Y axis', rotation=0, labelpad=20)
plt.title('title')

# loc : 범례의 위치 표시(upper, center, lower) + (left, center, right)
plt.legend(loc='upper right', prop={'size':7})
plt.show()
```

#### 바 차트(bar chart) : 막대 그래프

```python
# linewidth : 막대 테두리 굵기 설정
plt.bar(x, y, color=['cadetblue', 'cyan', 'lightskyblue'], edgecolor='blue', linewidth=4)
plt.show()
```

```python
# alpha : 투명도 설정
plt.bar(x, y, color=['cadetblue', 'cyan', 'lightskyblue'], edgecolor='blue', linewidth=4, alpha=0.5)
plt.show()
```

```python
# width : 막대 너비 설정
plt.bar(x, y, color=['cadetblue', 'cyan', 'lightskyblue'], 
        edgecolor='blue', linewidth=4, alpha=0.5, width=0.5)
plt.show()
```

#### 2) 요일별 사고 건수 시각화
- 요일별 사고 건수 확인
- 요일 순서대로 설정 후 변수에 저장해서 시각화에 활용

```python
# 요일별 사고 건수 확인
days = data['요일'].value_counts()  # 내림차순이 디폴트
```

```python
# 인덱스 자체만 바뀌고 값은 그대로
# days.index = ['월', '화', '수', '목', '금', '토', '일']
# days
```

#### 연습문제
### 사고유형이 차대차 사건 중에 죽거나 다친사람이 많은 발생지시도를 알아보고 시각화해보자!

```python
data['사고유형_대분류'].value_counts()
```

```python
data_ = data[data['사고유형_대분류']=='차대차'].loc[:,['사상자수', '발생지시도']].groupby('발생지시도').sum()
data_val = data_.values
data_lab = data_.index

plt.bar(np.arange(len(data_val)), data_val.reshape(-1))
plt.xticks(np.arange(len(data_val)), labels=data_lab, rotation=45)
plt.title('발생지시도 별 사상자수')
plt.xlabel('발생지시도')
plt.ylabel('사상자수')
plt.show()

```

```python
# 사고유형 대분류에서 차대차 사건만 불리언 인덱싱
ac = data[data['사고유형_대분류']=='차대차']
ac
```

```python
# 사고유형이 차대차인 사건은 총 1642건ㄴ
ac.shape
```

```python
ac = ac[['발생지시도', '사상자수']]
```

```python
ac['발생지시도'].value_counts()
```

```python
# 각 발생지시도 별로 사상자수의 합계 확인하기
# 그룹 별(발생지시도)로 묶어서 집계(사상자수의 합)를 출력
```

```python
ac2 = ac.groupby('발생지시도').sum()
ac2.sort()
ac2
```

```python
ac3 = ac2.sort_values('사상자수', ascending=False)
```

```python
fig = plt.figure(figsize=(10, 5))

plt.bar(np.arange(len(ac3)), ac3['사상자수'])

plt.xticks(np.arange(len(ac3)), labels=ac3.index)

plt.title('차vs차 교통사고의 사상자 수')
plt.xlabel('지역')
plt.ylabel('사상자 수')
plt.show()
```
