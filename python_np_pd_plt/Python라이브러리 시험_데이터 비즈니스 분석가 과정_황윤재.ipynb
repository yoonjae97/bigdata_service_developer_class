## 문제3

```python
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font', family = 'Malgun Gothic')

X = [2,3,4,5,6,7,8]
Y1 = [5,7,3,9,1,8,6]
Y2 = [9,5,7,2,3,7,1]

plt.plot(X, Y1, c='r', label='Y1')
plt.plot(X, Y2, c='b', label='Y2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('<실행결과>')
plt.legend()
plt.grid()
plt.show()
```

## 문제4

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import rc
# 조건 3
rc('font', family = 'Malgun Gothic')

data = pd.read_csv("Traffic_Accident_2017.csv", encoding='euc-kr')

# 조건 1
data_ = data[(data['사고유형_대분류']=='차대사람') & (data['주야']=='야간')].groupby('요일').count().iloc[:,0]

days = data_.reindex(['월', '화', '수', '목', '금', '토', '일'])

plt.bar(np.arange(len(days)), days.values)
plt.title('요일별 야간 차대사람 교통사고 건수')
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.xticks(np.arange(7), labels=days.index)
# 조건 2
plt.ylim(0, 200)
plt.show()
```
