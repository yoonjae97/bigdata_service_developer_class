### Pandas 데이터 분석 실습 예제
- 2015-2017년 광주광역시 범죄현황 데이터를 이용해 전년 대비 지역별 범죄 증감율 구하기

- 증감율 = (금년 - 작년)/작년 * 100
    - ex) 2015-16 증감율 = (15416 - 18830)/18830 * 100

```python
data_15 = pd.read_csv('2015.csv', encoding='euc-kr' )
data_16 = pd.read_csv('2016.csv', encoding='euc-kr')
data_17 = pd.read_csv('2017.csv', encoding='euc-kr')
```

```python
data_17 = data_17[data_17['관서명'] != '광주지방경찰청']
```

```python
data_15['2015총계'] = data_15.iloc[:, 2:].sum(axis=1)
data_16['2016총계'] = data_16.iloc[:, 2:].sum(axis=1)
data_17['2017총계'] = data_17.iloc[:, 2:].sum(axis=1)
```

```python
s15 = data_15[data_15.iloc[:,1] == '발생건수'][['관서명','2015총계']]
s16 = data_16[data_16.iloc[:,1] == '발생건수'][['관서명','2016총계']]
s17 = data_17[data_17.iloc[:,1] == '발생건수'][['관서명','2017총계']]
```

```python
s15.set_index('관서명', inplace=True)
s16.set_index('관서명', inplace=True)
s17.set_index('관서명', inplace=True)
```

```python
s1516 = (s16.values-s15.values)/s15.values * 100
s1617 = (s17.values-s16.values)/s16.values * 100
```

```python
s15['2015-16증감율'] = s1516
s16['2016-17증감율'] = s1617
```

```python
result = pd.concat([s15, s16, s17], axis=1)
result
```

```python
# 1. 데이터 로드
data_15 = pd.read_csv('2015.csv', encoding='euc-kr', index_col='관서명')
data_16 = pd.read_csv('2016.csv', encoding='euc-kr', index_col='관서명')
data_17 = pd.read_csv('2017.csv', encoding='euc-kr', index_col='관서명')
```

```python
# 2. 데이터 전처리
data_17.drop('광주지방경찰청', inplace=True)
```

```python
# 3. 지역별 범죄 증감율을 구하기 위해 범죄 발생 횟수의 총합 필요
data_15['총계'] = data_15.loc[:, '살인':].sum(axis=1)
data_16['총계'] = data_16.loc[:, '살인':].sum(axis=1)
data_17['총계'] = data_17.loc[:, '살인':].sum(axis=1)
```

```python
# 4. 데이터 불리언 인덱싱
s15 = data_15.loc[data_15['구분']=='발생건수']['총계']
s16 = data_16.loc[data_16['구분']=='발생건수']['총계']
s17 = data_17.loc[data_17['구분']=='발생건수']['총계']
```

```python
# 5. 증감율 게산
s1516 = (s16 - s15) / s15 * 100
s1617 = (s17 - s16) / s16 * 100
```

```python
# 6. 만들어진 Series들을 병합 후 결과 출력
result = pd.concat([s15, s1516, s16, s1617, s17], axis=1)
result.columns = ['2015총계', '2015-16증감율', '2016총계', '2016-17증감율', '2017총계']
result
```
