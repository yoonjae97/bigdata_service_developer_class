## 인사이트 도출
- Warehouse_block : 지역에 따른 창고라면 지역과 연관 있을 가능성 
- Mode_of_Shipment : 해외 or 국내배송 가능성
- Customer_care_calls : 배송 문의 전화를 많이 해서 일찍 올 수 있다?
- Customer_rating : 배송 후의 평점이라면 배송이 빨리 왔을 경우 평점이 높을 것으로 예상
- Cost_of_the_Product : 상품 가격과의 연관성은 커보이지 않음
- Prior_purchases : 연관성이 높다고 생각했으나 이전 구매 경험이 많은 사람들의 배송이 제때 도착하지 않은 것으로 보아 재고 가능성 있음
- Product_importance : 중요도가 높은 상품을 빨리 배송해줄 가능성이 있음
- Gender : 성별은 상관관계가 커보이지 않음
- Discount_offered : 재고가 얼마 없어 세일로 팔아버린 경우라면 배송이 느릴 수 있고, 재고가 많아 세일을 하는 경우라면 배송이 빠를 수 있음
- Weight_in_gms : 무게에 따른 배송 방법의 차이과 배송 창고가 달라지는 경우가 아니라면 상관관계는 없지만 재고 가능성 있음

```python
# 방법 1

gbc = create_model('gbc')
tuned_gbc = tune_model(gbc, search_library='optuna')
catboost = create_model('catboost')
tuned_catboost = tune_model(catboost, search_library='optuna')
rf = create_model('rf')
tuned_rf = tune_model(rf, search_library='optuna')
```

```python
blender_4 = blend_models(estimator_list = [tuned_gbc, tuned_rf])

final_model = finalize_model(blender_4)
```
